```python
import matplotlib.pyplot as plt

# Example data (three series with 7 values each)
x = [1, 2, 3, 4, 5, 6, 7]
y1 = [10, 12, 15, 18, 20, 150, 200]
y2 = [8, 11, 14, 17, 19, 140, 180]
y3 = [9, 13, 16, 19, 21, 160, 210]

# Create two subplots with shared x-axis
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(8, 6), gridspec_kw={'height_ratios': [1, 1]})

# Plot each series on both axes
ax1.plot(x, y1, marker='o', label="Series 1")
ax1.plot(x, y2, marker='s', label="Series 2")
ax1.plot(x, y3, marker='^', label="Series 3")

ax2.plot(x, y1, marker='o', label="Series 1")
ax2.plot(x, y2, marker='s', label="Series 2")
ax2.plot(x, y3, marker='^', label="Series 3")

# Set the y-axis limits for each subplot
ax1.set_ylim(100, 220)  # Focus on higher values
ax2.set_ylim(0, 25)    # Focus on lower values

# Hide spines to indicate breaks
ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)

# Add diagonal lines to indicate the breaks
d = 0.015  # Size of the diagonal lines
kwargs = dict(color='k', clip_on=False)

# Diagonal lines between ax1 and ax2
ax1.plot((-d, +d), (-d, +d), transform=ax1.transAxes, **kwargs)  # Top-left
ax1.plot((1 - d, 1 + d), (-d, +d), transform=ax1.transAxes, **kwargs)  # Top-right
ax2.plot((-d, +d), (1 - d, 1 + d), transform=ax2.transAxes, **kwargs)  # Bottom-left
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), transform=ax2.transAxes, **kwargs)  # Bottom-right

# Add labels, title, and legend
ax1.legend(loc='upper left')
ax2.set_xlabel('X-axis')
fig.text(0.04, 0.5, 'Y-axis', va='center', rotation='vertical')
fig.suptitle('Chart with Broken Y-Axis for Multiple Series')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
